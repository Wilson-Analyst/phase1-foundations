```python
#numpy array

prices = np.array([100, 250, 80, 430, 175])

tax = 0.15

discount = 0.10

prices_with_tax = prices + (prices * tax)

prices_with_discount = prices - (prices * discount)

print("Prices with tax:", prices_with_tax)
print("Prices with discount:", prices_with_discount)
print("Mean price with discount:", prices_with_discount.mean())
```

```python
import numpy as np
import time

data = np.random.randint(1, 100, size=1_000_000)

#loop version
def loop_square(arr):
    result = []
    for x in arr:
        result.append(x ** 2)
    return result

#Numpy version
def numpy_square(arr):
    return arr ** 2
```

```python
%timeit loop_square(data)
%timeit numpy_square(data)
```

```python
# square the data in an array of 1 million integers

data = np.random.randint(1, 100, size=1_000_000)

squared_data = data ** 2

print(squared_data)
```

```python
#mean of 1 million floata
floats = np.random.rand(1_000_000)

mean_value = floats.mean()

print("Mean of the floats:", mean_value)
```

```python
# add two arrays of 1 million numbers element by element
array1 = np.random.randint(1, 100, size=1_000_000)
array2 = np.random.randint(1, 100, size=1_000_000)

result = array1 + array2

print(result)
```

```python
#numbers greater than 500

numbers = np.random.randint(1, 1000, size=1_000_000)

filtered_numbers = numbers[numbers > 500]
print(filtered_numbers)
```

```python
# array normalization
data = np.random.randint(1, 100, size=1_000_000)

data_sub = data - data.mean()

data_normalized = data_sub / data.std()

print(data_normalized)
```
